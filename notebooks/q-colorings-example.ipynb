```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 4
p = 0.7
G = nx.erdos_renyi_graph(n, p, seed=None, directed=False)
```

```python
nx.draw(G, with_labels = True)
plt.show()
```

```python
#Setting attributes for G:
nx.set_node_attributes(G, {0: 1, 1: 2, 2: 3, 3: 4, 4: 5}, name = "color")
```

```python
def uniform_q_colorings(G, n, n_sample):
    X = [None]*n_sample
    X[0] = [G.nodes[i]["color"] for i in range(0,n)]
    
    for i in range(1, n_sample):
        v = np.random.randint(n, size=1)[0]
        neigh = [n for n in G.neighbors(v)]

        q = [1, 2, 3, 4, 5] #numbers of colors
        
        for u in neigh:
            color_q = G.nodes[u]["color"]
            if color_q not in q:
                q.remove(color_q)

        if len(q) > 0:
            new_q = np.random.choice(q)
            G.nodes[u]["color"] = new_q
            
        X[i] = [G.nodes[i]["color"] for i in range(0,n)]

    return X
```

```python
uniform_q_colorings(G, n, 10)
```
