```python
import pickle
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = False
```

```python
# Open file in read-binary mode
with open('data/samples/verification_test_simple_path.pickle', 'rb') as file:
    # Load the pickled data
    data = pickle.load(file)
```

Plottting the probability $Pr(t_v < \infty| D_{obs}, \beta)$ as it accumulates

```python
def cummulative_probabilities(node, tree_list):
    prob = []
    counting = 0
    N = 0

    for tree in tree_list:
        N += 1
        if node in tree:counting += 1
        
        prob.append(counting/N)

    return prob
```

```python
it_1000 = cummulative_probabilities(2, data[0][0:1000])
```

Ploting the probability as it accumulates for a fixed number of iterations $n = 1000$.

```python
plt.plot(it_1000, label=r'$n = 1000$')
plt.ylabel(r'$Pr(t_v < \infty | D_{obs}, \beta)$')
plt.legend()
plt.show()
```
